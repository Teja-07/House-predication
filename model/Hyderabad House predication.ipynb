```python
df1['price_per_sqft']=df1['Price']/df1['Area']
```

```python
df1.head()
```

```python
df1.Location=df1.Location.apply(lambda x:x.strip())
location_stats=df1.Location.value_counts(ascending=True)
location_stats
```

```python
df2=df1[~(df1['Area']/df1['Bedrooms']<350)]
df2.shape
```

```python
df2.head()
```

```python
#here we are filtering beyond the std and also we are finding the price per sqft which is above then std
import numpy as np
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('Location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
```

```python
df3=remove_pps_outliers(df2)
df3.shape
```

```python
df3.price_per_sqft.describe()
```

```python
def plot_scatter_chart(df,Location):
    bhk2=df[(df.Location==Location) & (df.Bedrooms==2)]
    bhk3=df[(df.Location==Location) & (df.Bedrooms==3)]
    plt.scatter(bhk2.Area,bhk2.Price,marker='+',label='2 BHK')
    plt.scatter(bhk3.Area,bhk3.Price,label='3 BHK')
    plt.xlabel("Total area per sqft")
    plt.ylabel("Total amount")
    plt.legend()
plot_scatter_chart(df3,"Adibatla")
```

```python
#In histogram we pass only one x values and y will be the count
plt.hist(df3.price_per_sqft,rwidth=0.8)
plt.xlabel('price per sqft')
plt.ylabel('count')
plt.show()
```

```python
dummies=pd.get_dummies(df3.Location)
dummies
```

```python
df4=pd.concat([df3,dummies],axis='columns')
df5=df4.drop('Location',axis='columns')
df5.head()
```

```python
df5.shape
```

```python
X=df5.drop('Price',axis='columns')
X.head()
```

```python
y=df5.Price
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lg=LinearRegression()
lg.fit(X_train,y_train)
lg.score(X_train,y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
Dt=DecisionTreeRegressor()
Dt.fit(X_train,y_train)
Dt.score(X_train,y_train)
```

```python
from sklearn.linear_model import Lasso
ls=Lasso()
ls.fit(X_train,y_train)
ls.score(X_train,y_train)
```

```python
def predict_price(location,sqft,bhk):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return lg.predict([x])[0]
```
